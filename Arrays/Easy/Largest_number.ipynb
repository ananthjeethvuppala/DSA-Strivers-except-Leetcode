```python
arr = [10, 9, 7, 8, 1, 5]
```

### 🔹 Method 1: Using Built-in max()

```python
largest = max(arr)
print(f"Largest Number is {largest}")
```

### 🔹 Method 2: Using Loop (Step by Step)

```python
largest =  arr[0]
for num in arr:
    if largest < num:
        largest = num

print(f"Largest number is {largest}")
```

### 🔹 Method 3: Using Sorting

```python
arr.sort()
print(f"Largest number is: {arr[-1]}")
```
